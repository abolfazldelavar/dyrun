## 2D Line plot

The below part of code is written to illustrate 2D line graphs.

```python
# Setting the font of LaTeX
plt.rcParams.update({
    "text.usetex": True,
    "font.family": ""       # Times New Roman, Helvetica
})

# To set a size for all figures
plt.rcParams["figure.figsize"] = [7.8, 5.6]

# Set all font properties
plt.rc('font', size = 17)
# Set all plotted lines properties
plt.rc('lines', linewidth = 0.5)
# plt.rc('spines', linewidth = 0.5)

# Change the color of all axes, ticks and grid
AXES_COLOR = '#111'
mpl.rc('axes', edgecolor = AXES_COLOR, linewidth=0.5, labelcolor=AXES_COLOR)
mpl.rc('xtick', color = AXES_COLOR, direction='in')
mpl.rc('ytick', color = AXES_COLOR, direction='in')
mpl.rc('grid', color = '#eee')

# Data
t  = np.linspace(0.0, 2.0, 1000)
s1 = np.cos(4 * np.pi * t*t)/(t + 1) + 2
s2 = np.sin(4 * np.pi * t) + 2 + np.random.normal(0, 0.2, [1000,])

# Starting to creat and plot
h1  = plt.figure(figsize=(7.8, 5.6), tight_layout=True)
ax = h1.subplots()

# The main part of plots
# Alpha is transparent parameter
# mec denotes the transparent amount of circles borders
ax.plot(t, s1, 'o-', label = 'Eq. 1, $xy^2$', markevery=50, mec='0.7')
ax.plot(t, s2, ':k', label = 'Eq. 2 $\sqrt{\pi^2r}$', alpha=0.5)

# Hide the right and left
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)

# Set the others width thinner
# Because of initial options, it does not need to this now.
# ax.spines.bottom.set_linewidth(0.5)
# ax.spines.left.set_linewidth(0.5)

# Other options
ax.set_xlabel(r'time (s)', color='#888')
ax.set_ylabel(r'Amplitude')
ax.set_title(r'\TeX\ is the umber of $\displaystyle\sum_{n=1}^\infty'
             r'\frac{-e^{i\pi}}{2^n}$!', fontsize=12)
ax.set_xlim(0,2)
ax.set_ylim(0,4)

# How many ticks do you want to have in each axis
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

# Legend options
# 'frameon' can hide the background and the border of the legend
# 'fontsize' changes the font. The value must be True or False
# 'bbox_to_anchor' can change the position of the legend arbitrary
# 'ncol' indicated the number of columns that components are arranged.
# 'mode' implies the mode of legend whcih can be 'expand' or ?
ax.legend(loc='upper left', frameon=False, fontsize=13, bbox_to_anchor=(0,1,1,0), ncol=2, mode='expand')

# All ticks control
# In the initials, all have done.
# ax.tick_params(colors='y', width=0.5, direction='in')

# Set the color of thicks and labels 
ax.xaxis.set_tick_params(colors='r', color='b', pad=10)
ax.yaxis.set_tick_params(color='b', pad=10)

# The distance between labels and their axis
ax.xaxis.labelpad = 7
ax.yaxis.labelpad = 10

# To invert a axis
ax.invert_yaxis()

# A super title for the whole figure
h1.suptitle('Hi there')

# To save, Use the below code
# h.savefig('data/outputs/test.pdf', transparent=True, dpi=300)
# h.savefig('data/outputs/test.png', transparent=True, dpi=300)
# h.savefig('data/outputs/test.svg', transparent=True, dpi=300)

# Make it visible
plt.show()
```

## 3D Line plotting

The below part of code can be help to make 3D curves.

```python

# Setting the font of LaTeX
plt.rcParams.update({
    "text.usetex": True,
    "font.family": ""       # Times New Roman, Helvetica
})

# Set all font properties
plt.rc('font', size = 17)
# Set all plotted lines properties
plt.rc('lines', linewidth = 1)
# plt.rc('spines', linewidth = 0.5)

# Data
t  = np.linspace(0.0, 2.0, 1000)
s1 = np.cos(4 * np.pi * t*t)/(t + 1) + 2
s2 = np.sin(4 * np.pi * t) + 2 + np.random.normal(0, 0.2, [1000,])
s3 = np.cos(2 * np.pi * t + 2) + 2 + np.random.normal(0, 0.1, [1000,])


# Starting to creat and plot
h2  = plt.figure(figsize=(7.8, 5.5), tight_layout=True)
# This order can prepare the are for 3D curve plotting
ax = h2.add_subplot(projection='3d')

# The main part of plots
# Alpha is transparent parameter
# mec denotes the transparent amount of circles borders
ax.plot(t, s1, s2, label = 'Eq. 1, $xy^2$')
ax.plot(t, s1, s3, ':k', label = 'Eq. 2 $\sqrt{\pi^2r}$', alpha=0.5)

# Hide the right and left
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)

# Set the others width thinner
ax.spines.bottom.set_linewidth(0.5)
ax.spines.left.set_linewidth(0.5)

# Other options
ax.set_xlabel(r'time (s)', color='#888')
ax.set_ylabel(r'd$x$/d$t$')
ax.set_zlabel(r'Amplitude')
ax.set_xlim(0,2)
ax.set_ylim(0,4)

# How many ticks do you want to have in each axis
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.zaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

# Legend options
# 'frameon' can hide the background and the border of the legend
# 'fontsize' changes the font. The value must be True or False
# 'bbox_to_anchor' can change the position of the legend arbitrary
# 'ncol' indicated the number of columns that components are arranged.
# 'mode' implies the mode of legend whcih can be 'expand' or ?
ax.legend(loc='upper left', frameon=False, fontsize=13, bbox_to_anchor=(0,1,1,0), ncol=2, mode='expand')

# The view of camera (degree)
ax.view_init(20, -40)

# Set the color of thicks and labels 
ax.w_xaxis.set_tick_params(colors='r', color='none')
ax.w_yaxis.set_tick_params(color='none')
ax.w_zaxis.set_tick_params(color='none')

# Change the color of axis panes
# ax.w_xaxis.set_pane_color((1, 1, 1, 1))
# ax.w_yaxis.set_pane_color((1, 1, 1, 1))
# ax.w_zaxis.set_pane_color((1, 1, 1, 1))
# Or:
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False


# Grid lines
# Because of initial options, it does not need to this now.
# ax.w_xaxis._axinfo["grid"].update({"linewidth":0.5, "color" : '#e6e6e6'})
# ax.w_yaxis._axinfo["grid"].update({"linewidth":0.5, "color" : '#e6e6e6'})
# ax.w_zaxis._axinfo["grid"].update({"linewidth":0.5, "color" : '#e6e6e6'})

# The distance between labels and their axis
ax.xaxis.labelpad = 10
ax.yaxis.labelpad = 10
ax.zaxis.labelpad = 10

# If you want the graph without cordinates, use:
# ax.set_axis_off()

# Make it visible
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import show

# Set up a default, sample figure. 
fig = plt.figure()
x = np.linspace(-np.pi,np.pi,100)
y = 2*np.sin(x)

ax = fig.add_subplot(1,2,2)
ax.plot(x,y)
ax.set_title('Normal Spines')

def hide_spines():
    """Hides the top and rightmost axis spines from view for all active
    figures and their respective axes."""

    # Retrieve a list of all current figures.
    figures = [x for x in matplotlib._pylab_helpers.Gcf.get_all_fig_managers()]
    for figure in figures:
        # Get all Axis instances related to the figure.
        for ax in figure.canvas.figure.get_axes():
            # Disable spines.
            ax.spines['right'].set_color('none')
            ax.spines['top'].set_color('none')
            # Disable ticks.
            ax.xaxis.set_ticks_position('bottom')
            ax.yaxis.set_ticks_position('left')

hide_spines()
show()
```

```python
fig, ax = plt.subplots()
# interface tracking profiles
N = 500
delta = 0.6
X = np.linspace(-1, 1, N)
ax.plot(X, (1 - np.tanh(4 * X / delta)) / 2,    # phase field tanh profiles
        X, (1.4 + np.tanh(4 * X / delta)) / 4, "C2",  # composition profile
        X, X < 0, "k--")                        # sharp interface

# legend
ax.legend(("phase field", "level set", "sharp interface"),
          shadow=True, loc=(0.01, 0.48), handlelength=1.5, fontsize=16)

# the arrow
ax.annotate("", xy=(-delta / 2., 0.1), xytext=(delta / 2., 0.1),
            arrowprops=dict(arrowstyle="<->", connectionstyle="arc3"))
ax.text(0, 0.1, r"$\delta$",
        color="black", fontsize=24,
        horizontalalignment="center", verticalalignment="center",
        bbox=dict(boxstyle="round", fc="white", ec="black", pad=0.2))

# Use tex in labels
ax.set_xticks([-1, 0, 1])
ax.set_xticklabels(["$-1$", r"$\pm 0$", "$+1$"], color="k", size=20)

# Left Y-axis labels, combine math mode and text mode
ax.set_ylabel(r"\bf{phase field} $\phi$", color="C0", fontsize=20)
ax.set_yticks([0, 0.5, 1])
ax.set_yticklabels([r"\bf{0}", r"\bf{.5}", r"\bf{1}"], color="k", size=20)

# Right Y-axis labels
ax.text(1.02, 0.5, r"\bf{level set} $\phi$",
        color="C2", fontsize=20, rotation=90,
        horizontalalignment="left", verticalalignment="center",
        clip_on=False, transform=ax.transAxes)

# Use multiline environment inside a `text`.
# level set equations
eq1 = (r"\begin{eqnarray*}"
       r"|\nabla\phi| &=& 1,\\"
       r"\frac{\partial \phi}{\partial t} + U|\nabla \phi| &=& 0 "
       r"\end{eqnarray*}")
ax.text(1, 0.9, eq1, color="C2", fontsize=18,
        horizontalalignment="right", verticalalignment="top")

# phase field equations
eq2 = (r"\begin{eqnarray*}"
       r"\mathcal{F} &=& \int f\left( \phi, c \right) dV, \\ "
       r"\frac{ \partial \phi } { \partial t } &=& -M_{ \phi } "
       r"\frac{ \delta \mathcal{F} } { \delta \phi }"
       r"\end{eqnarray*}")
ax.text(0.18, 0.18, eq2, color="C0", fontsize=16)

ax.text(-1, .30, r"gamma: $\gamma$", color="r", fontsize=20)
ax.text(-1, .18, r"Omega: $\Omega$", color="b", fontsize=20)

plt.show()
```

```python
import proplot as pplt
labels = list('xyzpq')
state = np.random.RandomState(51423)
fig = pplt.figure(share=0, refwidth=2.3, suptitle='On-the-fly colorbar and legend demo')

# Legends
data = (state.rand(30, 10) - 0.5).cumsum(axis=0)
ax = fig.subplot(121, title='On-the-fly legend')
ax.plot(  # add all at once
    data[:, :5], lw=2, cycle='Reds1', cycle_kw={'ls': ('-', '--'), 'left': 0.1},
    labels=labels, legend='b', legend_kw={'title': 'legend title'}
)
for i in range(5):
    ax.plot(  # add one-by-one
        data[:, 5 + i], label=labels[i], linewidth=2,
        cycle='Blues1', cycle_kw={'N': 5, 'ls': ('-', '--'), 'left': 0.1},
        colorbar='ul', colorbar_kw={'label': 'colorbar from lines'}
    )

# Colorbars
ax = fig.subplot(122, title='On-the-fly colorbar')
data = state.rand(8, 8)
ax.contourf(
    data, cmap='Reds1', extend='both', colorbar='b',
    colorbar_kw={'length': 0.8, 'label': 'colorbar label'},
)
ax.contour(
    data, color='gray7', lw=1.5,
    label='contour', legend='ul', legend_kw={'label': 'legend from contours'},
)
```

## Image illustrator
This part can be used for imagination

```python
delta = 0.5
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z  = (Z1 - Z2) * 2

fig, ax = plt.subplots()
im = ax.imshow(Z, 
        interpolation = 'bilinear', # nearest, bilinear, bicubic
        cmap   = mpl.cm.RdYlGn,
        origin = 'lower',           # This can reverse the Y axis
        # extent = [0, 40, 0, 20],
        vmax   = abs(Z).max(),      # The maximum value
        vmin   = -abs(Z).max())     # The mimimum value

plt.yticks([0, 2, 4, 8, 10], ['a0', 'a2', 'a4', 'a8', 'a10'])
plt.colorbar(im, fraction =0.046, pad = 0.02)
plt.show()
```
